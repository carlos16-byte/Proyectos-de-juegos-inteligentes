### üìò Reglas b√°sicas del puzzle

1. Cada rompecabezas contiene una **rejilla de bloques seleccionables**.
2. El rompecabezas comienza seleccionando un **bloque que tenga un n√∫mero**.
3. Ese n√∫mero indica cu√°ntos **bloques conectados** (adyacentes) deben seleccionarse para completar el grupo.
4. Una vez completado un grupo, el siguiente bloque seleccionado debe contener otro n√∫mero.
5. Los bloques seleccionados siempre deben estar conectados al bloque anterior.
6. El rompecabezas se resuelve cuando **todos los bloques han sido seleccionados**.
7. Algunos puzzles pueden tener **m√∫ltiples soluciones**; los m√°s dif√≠ciles solo una.
8. Si seleccionas un bloque inv√°lido, este se pondr√° **rojo**.
9. El √∫ltimo bloque de un grupo debe estar conectado a un bloque con n√∫mero.
10. Si te atascas, puedes **empezar desde otro bloque numerado**.

---

### üìñ Tutorial paso a paso

* Paso 1: Selecciona un bloque que contenga un n√∫mero (ejemplo: el 2).
* Paso 2: Como seleccionaste el 2, selecciona **2 bloques m√°s conectados**.
* Paso 3: Ahora selecciona un bloque con n√∫mero que est√© junto al grupo anterior.
* Paso 4: Si es un 3, selecciona **3 bloques m√°s conectados**.
* Paso 5: Repite este proceso hasta que todos los bloques est√©n seleccionados.

---

### ‚ö° Conceptos especiales

* **Soluciones √∫nicas**: Los niveles f√°ciles pueden tener varias soluciones, pero los dif√≠ciles solo una.
* **√öltimo bloque seleccionado**: Algunos bloques deben ser el √∫ltimo del puzzle (ejemplo: un bloque con 0 que no est√© junto a otro bloque numerado).
* **Primer bloque seleccionado**: Algunos bloques "atrapados" deben seleccionarse primero porque de lo contrario no se podr√≠an alcanzar.
* **Uso de bloques con 0**: Sirven como puente para acceder a bloques numerados adyacentes.
* **Bloques inaccesibles**: Hay bloques que solo pueden alcanzarse desde ciertos n√∫meros; reconocer esto ayuda a planear la estrategia.

# Numari

````python
class NumariGame:
    def __init__(self, size=4):
        self.size = size  # Tama√±o del tablero (4x4, 5x5 o 6x6)
        self.levels = self._create_levels()  # Diccionario con niveles predefinidos
        self.board = None  # Matriz del tablero actual
        self.selected = None  # Matriz para marcar celdas seleccionadas
        self.current_level = 0  # Nivel actual
        self.moves = 0  # Contador de movimientos
        self.to_select = 0  # Celdas restantes por seleccionar
        self.expect_number = True  # Controla si se espera n√∫mero o celda vac√≠a
        self.last_pos = None  # √öltima celda seleccionada
        self.player_pos = None  # Posici√≥n actual del jugador

    def _create_levels(self):
        # Crea niveles seg√∫n el tama√±o del tablero
        if self.size == 4:
            return {
                1: np.array([
                    [0, 0, 0, 3],
                    [3, 0, 3, 0],
                    [0, 0, 0, 0],
                    [0, 0, 3, 0]
                ]),
                2: np.array([
                    [0, 4, 0, 0],
                    [0, 0, 0, 2],
                    [3, 0, 0, 0],
                    [0, 0, 0, 3]
                ]),
                3: np.array([
                    [0, 3, 0, 2],
                    [0, 0, 1, 0],
                    [0, 0, 0, 0],
                    [3, 0, 0, 2]
                ])
            }
        elif self.size == 5:
            return {
                1: np.array([
                    [3, 0, 0, 0, 0],
                    [0, 0, 0, 3, 0],
                    [0, 0, 3, 0, 0],
                    [0, 3, 0, 0, 0],
                    [3, 0, 0, 0, 4]
                ]),
                2: np.array([
                    [0, 0, 3, 0, 1],
                    [0, 3, 0, 0, 0],
                    [2, 0, 0, 0, 2],
                    [0, 3, 0, 0, 2],
                    [5, 0, 1, 0, 0]
                ]),
                3: np.array([
                    [0, 1, 0, 3, 0],
                    [2, 0, 0, 0, 0],
                    [2, 0, 2, 0, 2],
                    [0, 1, 0, 1, 0],
                    [5, 0, 0, 0, 0]
                ])
            }
        elif self.size == 6:
            return {
                1: np.array([
                    [4, 0, 0, 0, 4, 0],
                    [0, 0, 0, 0, 2, 0],
                    [0, 4, 0, 0, 0, 0],
                    [0, 3, 0, 0, 0, 3],
                    [0, 0, 0, 2, 0, 0],
                    [0, 1, 0, 4, 0, 0]
                ]),
                2: np.array([
                    [0, 1, 0, 0, 2, 0],
                    [0, 0, 0, 3, 0, 0],
                    [2, 1, 0, 1, 0, 2],
                    [0, 0, 1, 0, 3, 0],
                    [0, 0, 0, 0, 0, 0],
                    [0, 3, 2, 0, 3, 0]
                ]),
                3: np.array([
                    [3, 0, 4, 0, 0, 0],
                    [0, 0, 2, 0, 3, 0],
                    [0, 0, 0, 0, 0, 0],
                    [0, 4, 0, 0, 1, 0],
                    [0, 0, 0, 3, 0, 0],
                    [3, 0, 0, 0, 4, 0]
                ])
            }
        return {}  # Si no coincide tama√±o, retorna vac√≠o

    def load_level(self, level_num):
        # Carga un nivel espec√≠fico
        if level_num in self.levels:
            self.current_level = level_num  # Guarda n√∫mero de nivel
            self.board = self.levels[level_num].copy()  # Copia tablero
            self.selected = np.full_like(self.board, False, dtype=bool)  # Reinicia selecci√≥n
            self.moves = 0  # Reinicia movimientos
            self.to_select = 0  # Reinicia contador de celdas por llenar
            self.expect_number = True  # Vuelve a esperar n√∫mero
            self.last_pos = None  # Limpia √∫ltima posici√≥n
            self.player_pos = None  # Limpia posici√≥n jugador
            return True
        return False  # Nivel inv√°lido

    def display_board(self):
        # Muestra el tablero en consola
        print("="*50)
        print(f"NUMARI - Matriz {self.size}x{self.size} - Nivel {self.current_level}")
        print(f"Movimientos: {self.moves}")
        print("="*50)

        # Mostrar encabezado con columnas
        print("    ", end="")
        for col in range(self.size):
            print(f"{col:^4}", end="")
        print()

        # Mostrar filas con contenido
        for r in range(self.size):
            print(f"{r:^3} ", end="")
            for c in range(self.size):
                val = self.board[r, c]
                if self.selected[r, c]:  # Si la celda est√° seleccionada
                    print(f"\033[0;32m[\033[0m{val if val > 0 else '¬∑'}\033[0;32m]\033[0m", end=" ")
                else:
                    print(f" {val if val > 0 else '¬∑'} ", end=" ")
            print()
        print("="*50)

    def is_adjacent(self, r, c):
        # Comprueba si la celda es adyacente a la √∫ltima seleccionada
        if self.last_pos is None:
            return True
        lr, lc = self.last_pos
        return abs(lr - r) + abs(lc - c) == 1

    def select_cell(self, r, c):
        # Intenta seleccionar una celda del tablero
        if not (0 <= r < self.size and 0 <= c < self.size):
            print("Posici√≥n fuera del tablero.")
            return False

        if self.selected[r, c]:
            print("Esa celda ya est√° seleccionada.")
            return False

        if not self.is_adjacent(r, c):
            print("Debes seleccionar una celda adyacente al bloque anterior.")
            return False

        val = self.board[r, c]  # Valor de la celda

        if self.expect_number:
            # Se espera seleccionar un n√∫mero
            if val <= 0:
                print("Debes seleccionar un bloque con un n√∫mero.")
                return False
            self.to_select = val  # Guardar n√∫mero de celdas a llenar
            self.expect_number = False  # Cambiar estado
            print(f"Has seleccionado un n√∫mero {val}. Ahora selecciona {val} celdas vac√≠as conectadas.")
        else:
            # Se esperan celdas vac√≠as
            if val > 0:
                if self.to_select != 0:
                    print(f"A√∫n te faltan {self.to_select} celdas vac√≠as para completar el grupo.")
                    return False
                self.expect_number = True
                print(f"Has completado un grupo. Ahora selecciona un bloque con n√∫mero.")
            else:
                self.to_select -= 1  # Resta celdas por llenar
                print(f"Celdas restantes para este grupo: {self.to_select}")
                if self.to_select == 0:
                    self.expect_number = True
                    print("Grupo completado. Selecciona un bloque con n√∫mero.")

        self.selected[r, c] = True  # Marcar celda como usada
        self.last_pos = (r, c)  # Actualizar √∫ltima posici√≥n
        self.moves += 1  # Contar movimiento
        return True

    def is_completed(self):
        # Verifica si todo el tablero est√° cubierto
        return np.all(self.selected)

    def play(self):
        # L√≥gica principal del juego
        print("¬°Bienvenido a NUMARI!")
        while True:
            try:
                level = int(input("Selecciona nivel: "))  # Pedir nivel
                if self.load_level(level):
                    break
                else:
                    print("Nivel inv√°lido.")
            except ValueError:
                print("Entrada inv√°lida. Debe ser un n√∫mero.")

        # Selecci√≥n inicial con coordenadas
        while True:
            self.display_board()
            try:
                r = int(input("Selecciona fila inicial: "))
                c = int(input("Selecciona columna inicial: "))
                if self.select_cell(r, c):
                    self.player_pos = (r, c)
                    break
                else:
                    print("Posici√≥n inicial inv√°lida.")
            except ValueError:
                print("Entrada inv√°lida. Ingresa n√∫meros v√°lidos.")

        # Bucle principal de movimientos
        while True:
            self.display_board()
            if self.is_completed():
                print("¬°Felicidades, completaste el nivel!")
                break

            move = input("Escribe una acci√≥n (W/A/S/D = mover, Q = reiniciar, X = salir): ").lower().strip()

            if move == 'q':
                print("Reiniciando nivel...")
                self.load_level(self.current_level)
                # Pedir nueva celda inicial
                while True:
                    self.display_board()
                    try:
                        r = int(input("Selecciona fila inicial: "))
                        c = int(input("Selecciona columna inicial: "))
                        if self.select_cell(r, c):
                            self.player_pos = (r, c)
                            break
                        else:
                            print("Posici√≥n inicial inv√°lida.")
                    except ValueError:
                        print("Entrada inv√°lida. Ingresa n√∫meros v√°lidos.")
                continue

            if move == 'x':
                print("Saliendo del juego... ¬°Hasta la pr√≥xima!")
                return

            if move not in ['w', 'a', 's', 'd']:
                print("Acci√≥n inv√°lida. Usa solo W, A, S, D, Q o X.")
                continue

            row, col = self.player_pos
            new_row, new_col = row, col

            if move == 'w':
                new_row -= 1
            elif move == 's':
                new_row += 1
            elif move == 'a':
                new_col -= 1
            elif move == 'd':
                new_col += 1

            # Intentar seleccionar la nueva celda
            if 0 <= new_row < self.size and 0 <= new_col < self.size:
                if self.select_cell(new_row, new_col):
                    self.player_pos = (new_row, new_col)
                else:
                    print("Movimiento inv√°lido.")
            else:
                print("Movimiento fuera del tablero.")
````

````python
class NumariGameManager:
    def __init__(self):
        # Diccionario con instancias del juego para tableros 4x4, 5x5 y 6x6
        self.games = {4: NumariGame(4), 5: NumariGame(5), 6: NumariGame(6)}
        # Tama√±o de tablero por defecto (4x4)
        self.current_size = 4

    def select_board_size(self):
        # Permite al jugador elegir el tama√±o del tablero
        while True:
            try:
                print("Selecciona el tama√±o del tablero:")
                print("1. 4x4")
                print("2. 5x5")
                print("3. 6x6")
                choice = int(input("Nivel (1-3): "))  # Pedir elecci√≥n
                if choice == 1:
                    self.current_size = 4; break
                elif choice == 2:
                    self.current_size = 5; break
                elif choice == 3:
                    self.current_size = 6; break
                else:
                    print("Nivel inv√°lido")  # Validar que el nivel exista
            except ValueError:
                print("Debes ingresar un n√∫mero v√°lido")  # Asegurarse de que se ingrese un n√∫mero

    def start_game(self):
        # Inicia el flujo del juego seleccionando primero el tama√±o
        self.select_board_size()
        # Llama al m√©todo play() del tama√±o de tablero elegido
        self.games[self.current_size].play()

# Crear un administrador del juego
manager = NumariGameManager()
````

#Jugar Numari


````python
# Inicia Numari :)s
manager.start_game()
````
